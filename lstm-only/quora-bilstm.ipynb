```python
import tensorflow as tf
def bidirectional_layer(X, n_neurons):
    with tf.variable_scope("1", default_name="1", reuse=tf.AUTO_REUSE):
        cell_fw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        cell_bw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        outputs, output_states = tf.nn.bidirectional_dynamic_rnn(cell_fw, cell_bw, X, dtype=tf.float32)
        outputs = tf.concat(outputs, 2)
    return (outputs, output_states)
def bidirectional_layer_2(X, n_neurons):
    with tf.variable_scope("2", default_name="2", reuse=tf.AUTO_REUSE):
        cell_fw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        cell_bw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        outputs, output_states = tf.nn.bidirectional_dynamic_rnn(cell_fw, cell_bw, X, dtype=tf.float32)
        outputs = tf.concat(outputs, 2)
    return (outputs, output_states)
def bidirectional_layer_3(X, n_neurons):
    with tf.variable_scope("3", default_name="3", reuse=tf.AUTO_REUSE):
        cell_fw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        cell_bw = tf.nn.rnn_cell.LSTMCell(num_units=n_neurons)
        outputs, output_states = tf.nn.bidirectional_dynamic_rnn(cell_fw, cell_bw, X, dtype=tf.float32)
        outputs = tf.concat(outputs, 2)
    return (outputs, output_states)

MAX_LENGTH = 50
learning_rate = 0.01

def get_mini_batch(GloveEmbedding, train_data_frame, block_number1, block_number2, block_size):
    print("block_number-",block_number1,"--",block_number2)
    global MAX_LENGTH, EMBEDDING_DIM
    num_of_blocks = block_number2 - block_number1 + 1
    text_mini_batch = np.zeros((block_size*num_of_blocks, MAX_LENGTH, EMBEDDING_DIM))
    y_mini_batch = np.zeros((block_size*num_of_blocks, 1))
    mini_data_frame = train_data_frame.iloc[[i for i in range(block_size*(block_number1-1), block_number2*block_size)]]
    for index, row in mini_data_frame.iterrows(): 
        i = index - block_size*(block_number1-1)
        words = row['question_text'].strip().split(" ")
        for j in range(MAX_LENGTH):
            if j < len(words) and words[j] in GloveEmbedding:
                text_mini_batch[i][j] = GloveEmbedding[words[j]]

        y_mini_batch[i] = float(row['target'])
    return (text_mini_batch, y_mini_batch)

def get_test_batch(GloveEmbedding, test_data_frame, count):
    
    global MAX_LENGTH, EMBEDDING_DIM
    length = int(test_data_frame.size/2)
    #y_mini_batch = np.zeros((block_size*num_of_blocks, 1))
    #mini_data_frame = train_data_frame.iloc[[i for i in range(block_size*(block_number1-1), block_number2*block_size)]]
    start = int(length * count / 10)
    end = int(length * (count+1) / 10)
    if count == 9:
        end = length
    
    current_size = int(end - start)
    text_mini_batch = np.zeros((current_size, MAX_LENGTH, EMBEDDING_DIM))
    for index, row in test_data_frame.iterrows(): 
        if index >= start and index < end:
            words = row['question_text'].strip().split(" ")
            for j in range(MAX_LENGTH):
                if j < len(words) and words[j] in GloveEmbedding:
                    text_mini_batch[index-start][j] = GloveEmbedding[words[j]]
    mini_data_frame = test_data_frame.iloc[[i for i in range(start, end)]]
    return (text_mini_batch, mini_data_frame)

def correct_func(x, y):
    temp = (x > 0.5) == y
    return tf.cast(temp, tf.float32)

def f1_metric(logits, y):
    #element_wise_product = tf.math.multiply(logits, y)
    predicted = tf.cast(tf.math.greater(logits, 0.5), tf.float32)
    TP = tf.count_nonzero(predicted * y)
    TN = tf.count_nonzero((predicted - 1) * (y - 1))
    FP = tf.count_nonzero(predicted * (y - 1))
    FN = tf.count_nonzero((predicted - 1) * y)
    precision = TP / (TP + FP)
    recall = TP / (TP + FN)
    f1 = 2 * precision * recall / (precision + recall)
    return (precision, recall, f1)

def penalized_sigmoid(labels, logits):
    loss_vec = -5 * labels * tf.math.log(logits) - (1 - labels)*tf.math.log(1 - logits)
    return loss_vec
    
def get_submission_file(test_data_frame, logits_, outputf):
    
    logits_ = logits_ > 0.5
    logits_ = logits_ * 1
    
    print(logits_.shape, test_data_frame.size)
    for index, row in test_data_frame.iterrows():
        outputf.write(str(row["qid"]) + "," + str(logits_[index%logits_.shape[0]][0]) + "\n")
    
def set_values(test_data_frame, logits_, submission_data_frame, count):
    logits_ = logits_ > 0.5
    logits_ = logits_ * 1
    
    print(logits_.shape, test_data_frame.size)
    for index, row in test_data_frame.iterrows():
        submission_data_frame.loc[index] = [row["qid"], logits_[index%logits_.shape[0]][0]]
def model():
    global train_data_frame, MAX_LENGTH, EMBEDDING_DIM
    text = tf.placeholder(tf.float32, [None, MAX_LENGTH, EMBEDDING_DIM])
    y = tf.placeholder(tf.float32, [None, 1])
    outputs_layer_1, states_layer_1 = bidirectional_layer(text, 100)
    outputs_layer_2, states_layer_2 = bidirectional_layer_2(outputs_layer_1, 50)
    outputs_layer_3, states_layer_3 = bidirectional_layer_3(outputs_layer_2, 25)
    dimensions = []
    dimensions.append(tf.shape(outputs_layer_3)[0])
    dimensions.append(outputs_layer_3.get_shape()[1].value)
    dimensions.append(outputs_layer_3.get_shape()[2].value)
    
    flatten_vector = tf.reshape(outputs_layer_3, [dimensions[0], dimensions[1]*dimensions[2]])
    pre_final_vector = tf.contrib.layers.fully_connected(flatten_vector, 100, activation_fn= tf.nn.relu)
    logits = tf.contrib.layers.fully_connected(flatten_vector, 1, activation_fn=tf.nn.sigmoid)
    #xsigmoid = tf.nn.sigmoid_cross_entropy_with_logits(labels=tf.reshape(y,[tf.shape(y)[0], 1]), logits=logits)
    xsigmoid = penalized_sigmoid(labels=tf.reshape(y,[tf.shape(y)[0], 1]), logits=logits)
    loss = tf.reduce_mean(xsigmoid)
    optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
    training_op = optimizer.minimize(loss)
    # correct = tf.py_func(correct_func, [logits, y], tf.float32)
    precision, recall, f1 = f1_metric(logits, y)
    correct = tf.equal(tf.round(logits), y)
    accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))
    n_epochs = 10
    batch_size = 1000
    init = tf.global_variables_initializer()
    with tf.Session() as sess:
        init.run()
        num_examples = int(train_data_frame.size / 3)
        num_mini_batches = int(num_examples / batch_size)
        print("num_mini_batch",num_mini_batches)
        for epoch in range(n_epochs):
            length = int(num_mini_batches * 7 / 10)
            for iteration in range(length):
                text_mini_batch, y_mini_batch = get_mini_batch(GloveEmbedding, train_data_frame, iteration+1, iteration+1, batch_size)
                sess.run(training_op,feed_dict={text: text_mini_batch, y:y_mini_batch})
            text_mini_batch, y_mini_batch = get_mini_batch(GloveEmbedding, train_data_frame, int((0.9*num_mini_batches+length)/2), int((0.9*num_mini_batches+length)/2)+10, batch_size)
            #logits_ = logits.eval(feed_dict={text: text_mini_batch})
            #correct_ = correct.eval(feed_dict={text: text_mini_batch, y: y_mini_batch})
            #print(logits_)
            #print(y_mini_batch)
            #print(correct_)
            precision_train = precision.eval(feed_dict={text: text_mini_batch, y: y_mini_batch})
            recall_train = recall.eval(feed_dict={text: text_mini_batch, y: y_mini_batch})
            f1_train = f1.eval(feed_dict={text: text_mini_batch, y: y_mini_batch})
            accuracy_ = accuracy.eval(feed_dict={text: text_mini_batch, y: y_mini_batch})
            print(epoch, ",precision:", precision_train, ",recall:", recall_train,",f1:",f1_train,",accuracy:",accuracy_)
            #outputf = open("../submission.csv", "w")
            #outputf.write("qid" + "," + "prediction" + "\n")
        submission_data_frame = pd.DataFrame(columns=['qid', 'prediction'])
        for count in range(10):
            test_batch, mini_test_data_frame = get_test_batch(GloveEmbedding, test_data_frame, count)
            logits_ = logits.eval(feed_dict={text: test_batch})
            # get_submission_file(mini_test_data_frame, logits_, outputf)
            set_values(mini_test_data_frame, logits_, submission_data_frame, count)
            #print(submission_data_frame)
            #print(mini_test_data_frame)

        submission_data_frame.to_csv('submission.csv', index=False)
        #outputf.close()

model()
```
